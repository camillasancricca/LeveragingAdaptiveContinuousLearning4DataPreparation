```python
# Data for the imputation techniques and their corresponding frequencies
# Data for the imputation techniques and their corresponding frequencies
imputation_techniques = [
    "Standard", "Mean", "Median", "Random", "KNN", "MICE", "Linear Regression", "Random Forest", "C-means"
]

# Frequency of pulls per arm
train_pulls = [580., 591., 558., 547., 561., 492., 595., 220., 521.]
test_pulls = [36, 0, 19, 44, 0, 0, 21, 0, 0]

# Ground truth distribution
ground_truth = [11, 17, 16, 12, 19, 13, 11, 113, 14]

# Creating the plot
fig, ax = plt.subplots(figsize=(10, 6))
width = 0.3  # bar width

# Create bars for each set of data
ax.bar([x - width for x in range(len(train_pulls))], train_pulls, width=width, label='Train Pulls')
ax.bar(range(len(test_pulls)), test_pulls, width=width, label='Test Pulls')
ax.bar([x + width for x in range(len(ground_truth))], ground_truth, width=width, label='Ground Truth')

# Adding labels and title
ax.set_xlabel('Imputation Techniques', fontsize=12)
ax.set_ylabel('Frequency', fontsize=12)
ax.set_title('Comparison of Top-1 Imputation Technique Performance', fontsize=14)
ax.set_xticks(range(len(imputation_techniques)))
ax.set_xticklabels(imputation_techniques, rotation=45, ha='right')
ax.legend()

# Show plot
plt.tight_layout()
plt.show()
```
