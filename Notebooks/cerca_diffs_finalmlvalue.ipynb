```python
import pandas as pd
```

```python
og_dataset = pd.read_csv('datasetCreated/results.csv')
og_dataset.head()
```

```python
# For each unique triple data_object,initial_completeness, ml_algorith, calculate the percentage difference between top 3 max final_ml_value
# Group by the unique triple and find the top 3 max values for each group
grouped_dataset = og_dataset.groupby(['data_object', 'initial_completeness', 'ml_algorithm'])
```

```python
# Calculate the differences within the top k values
k = 3
result_diffs = {}
for name, group in grouped_dataset:
    top_k = group.nlargest(k, 'final_ml_value')['final_ml_value']
    differences = (top_k.max() - top_k)
    result_diffs[name] = differences
```

```python
result_diffs
```

```python
# Calculate the mean percentage difference for each unique triple
mean_diffs = {}
for key, value in result_diffs.items():
    mean_diffs[key] = value.mean()
```

```python
mean_diffs
```

```python
# Extract a list of the mean percentage differences
mean_diffs_list = []
for key, value in mean_diffs.items():
    mean_diffs_list.append(value)
```

```python
mean_diffs_list
```

```python
mean = sum(mean_diffs_list) / len(mean_diffs_list)
mean
```

# Diffs calculation in percentage 

```python
# Calculate the percentage difference for each unique triple
k = 3
result_percentage_diffs = {}
for name, group in grouped_dataset:
    top_k = group.nlargest(k, 'final_ml_value')['final_ml_value']
    differences = (top_k.max() - top_k) / top_k.max() * 100
    result_percentage_diffs[name] = differences
```

```python
result_percentage_diffs
```

```python
mean_percentage_diffs_list = []
for key, value in result_percentage_diffs.items():
    mean_percentage_diffs_list.append(value.mean())
```

```python
mean_percentage_diffs_list
```

```python
mean_percentage = sum(mean_percentage_diffs_list) / len(mean_percentage_diffs_list)
mean_percentage
```
